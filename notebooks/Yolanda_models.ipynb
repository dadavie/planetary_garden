```python
plantae=data.copy()
plantae=plantae[plantae['kingdom']=='Plantae']
plantae.shape
```

```python
preprocessed=plantae.copy()
data_strings=preprocessed[['Unnamed: 0.1', 'Unnamed: 0','gbifID','datasetKey','kingdom',
       'phylum','class','order','family','genus','species',
       'scientificName','verbatimScientificName','countryCode','locality',
       'stateProvince','decimalLatitude', 'decimalLongitude','year']]
preprocessed=preprocessed.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'gbifID', 'datasetKey', 'kingdom',
       'phylum', 'class', 'order', 'family', 'genus', 'species',
       'scientificName', 'verbatimScientificName', 'countryCode', 'locality',
       'stateProvince', 'decimalLatitude', 'decimalLongitude', 'year'])
```

```python
preprocessed
```

```python
from sklearn.preprocessing import Normalizer
scaler= Normalizer()
scaler = Normalizer().fit(preprocessed)  # fit does nothing.
preprocessed=scaler.transform(preprocessed)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
nrst_neigh = NearestNeighbors(n_neighbors = 10, algorithm='ball_tree')
nrst_neigh.fit(preprocessed)
distances, indices = nrst_neigh.kneighbors(preprocessed)
indices
distances
nrst_neigh.kneighbors_graph(preprocessed).toarray()
```
